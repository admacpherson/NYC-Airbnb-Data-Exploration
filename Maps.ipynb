```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))
nyc_map = gpd.read_file('neighbourhoods.geojson')
nyc_map
```

```python
bouroughs = nyc_map['neighbourhood_group'].value_counts().index
bouroughs
```

```python
# Create a figure and axis for the plot
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

# Plot the GeoDataFrame in color
nyc_map.plot(column='neighbourhood_group', ax=ax, legend=True, cmap='viridis')

# Save the color version
plt.savefig('nyc_neighbourhoods_map_color.png', dpi=300, bbox_inches='tight')

# Plot a black-and-white version
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
nyc_map.plot(column='neighbourhood_group', ax=ax, legend=True, cmap='Greys')

# Save the black-and-white version
plt.savefig('nyc_neighbourhoods_map_bw.png', dpi=300, bbox_inches='tight')

# Optionally display the plots
plt.show()
```

```python
# Get the bounding box of the GeoDataFrame
bounds = nyc_map.total_bounds

# Extract the coordinates
min_longitude, min_latitude, max_longitude, max_latitude = bounds

print(f"Min Longitude: {min_longitude}")
print(f"Min Latitude: {min_latitude}")
print(f"Max Longitude: {max_longitude}")
print(f"Max Latitude: {max_latitude}")
```
