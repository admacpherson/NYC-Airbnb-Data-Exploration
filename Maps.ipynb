```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))
nyc_map = gpd.read_file('neighbourhoods.geojson')
nyc_map
```

```python
bouroughs = nyc_map['neighbourhood_group'].value_counts().index
bouroughs
```

```python
# Plot the map, coloring by 'neighbourhood_group'
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
nyc_map.plot(column='neighbourhood_group', ax=ax, legend=True, cmap='viridis')

plt.savefig('nyc_neighborhoods_map.png', dpi=300, bbox_inches='tight')

# Display the plot
plt.show()

```
